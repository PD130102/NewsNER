##### Text Preprocessing and Feature Extraction

```python
from nltk.tokenize import word_tokenize
with open('./dataset/stopwords.txt', 'r') as f:
    stop_words = f.read().split('\n')
stop_words = set(stop_words)
print(stop_words)
```

```python
for i in range(len(text_data)):
    text_data[i] = text_data[i].replace('[^a-zA-Z0-9]', '')
```

```python
# Tokenization
tokenized_text = []
for i in range(len(text_data)):
    tokenized_text.append(word_tokenize(text_data[i]))
```

```python
tokenized_text[2]
```

```python
# Stopwords removal
filtered_text = []
for i in range(len(tokenized_text)):
    temp = []
    for j in range(len(tokenized_text[i])):
        if tokenized_text[i][j] not in stop_words:
            temp.append(tokenized_text[i][j])
    filtered_text.append(temp)
```

```python
filtered_text[2]
```

```python
for i in range(len(filtered_text)):
    for j in range(len(filtered_text[i])):
        # filtered_text[i][j] = re.sub(r'\W', '', filtered_text[i][j])
        filtered_text[i][j] = re.sub(r'\d+', '', filtered_text[i][j])
```

```python
# Remove . , ! ? `` '' etc
for i in range(len(filtered_text)):
    for j in range(len(filtered_text[i])):
        filtered_text[i][j] = re.sub(r'[.,!?;:/``]', '', filtered_text[i][j])
        filtered_text[i][j] = re.sub(r"''", '', filtered_text[i][j])
        filtered_text[i][j] = re.sub(r'``', '', filtered_text[i][j])
        
```

```python
filtered_text[2]
```

```python
# Remove words with length less than 2
for i in range(len(filtered_text)):
    temp = []
    for j in range(len(filtered_text[i])):
        if len(filtered_text[i][j]) > 1:
            temp.append(filtered_text[i][j])
    filtered_text[i] = temp
```

##### Stemming and Lemmatisation 

```python
import pandas as pd
from snowballstemmer import stemmer

hindi_stem = stemmer('hindi')

def hindi_stemming(word):
    return hindi_stem.stemWord(word)

def remove_stop_words(query, stop_words):
    return [word for word in query if word not in stop_words]

def search_in_dataset(query, dataset):
    return [word for word in query if word in dataset]
```

```python
def apply_suffix_rules(word):
    suffix_rules = [
        ("िय", "ी"), 
        ("ियां", "ी"), 
        ("िअों", "ी"), 
        ("यां", "ी"), 
        ("य", "ी"), 
        ("अों", ""), 
        ("ौं", ""), 
        ("े", ""), 
        ("ो", ""), 
        ("ों", ""), 
        ("ी", ""), 
        ("एँ", ""), 
        ("एं", ""), 
        ("ा", ""), 
        ("ं", ""), 
        ("ी", ""), 
        ("ए", "ा"), 
        ("ई", ""), 
        ("ना", ""), 
        ("ता", ""), 
        ("ती", ""), 
        ("जन", ""), 
        ("गण", ""), 
        ("तया", ""), 
        ("वग", ""), 
        ("कार", ""), 
        ("व", ""), 
        ("ें", ""), 
        ("ीय", ""), 
        ("़ियाँ", "या")
    ]
    
    for old_suffix, new_suffix in suffix_rules:
        if word.endswith(old_suffix):
            return word[:-len(old_suffix)] + new_suffix
    
    return word
```

```python
# Stemming 
for i in range(len(filtered_text)):
    for j in range(len(filtered_text[i])):
        # filtered_text[i][j] = hindi_stemming(filtered_text[i][j])
        filtered_text[i][j] = apply_suffix_rules(filtered_text[i][j])
```

##### N-Gram Model and Perplexity

```python
from nltk import ngrams
from collections import Counter

def get_ngrams(text, n):
    n_grams = ngrams(text, n)
    return [ ' '.join(grams) for grams in n_grams]

def get_ngrams_freq(text, n):
    n_grams = ngrams(text, n)
    return Counter(n_grams)

for i in range(len(filtered_text[:100])):
    print(get_ngrams(filtered_text[i], 2))
    print(get_ngrams_freq(filtered_text[i], 2))
```
